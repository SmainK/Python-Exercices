Input :

```python
df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
#Q1
cols = list(df.columns)
a, c = cols.index('a'), cols.index('c')
cols[c], cols[a] = cols[a], cols[c] 
df = df[cols]

#Q2

def swap_columns (df, col1, col2): 
    cols = list(df.columns)
    first, second = cols.index(col1), cols.index(col2)
    cols[second], cols[first] = cols[first], cols[second] 
    return df[cols]
    
test = swap_columns(df, 'c', 'a')
test

#Q3
#['c', 'b', 'a', 'd', 'e']
reverse = list(df.columns)[::-1]
df[reverse]

#Better solution : df.sort_index(axis=1, ascending=False, inplace=True)
```
